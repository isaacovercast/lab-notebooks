```python
%matplotlib inline

import itertools
import sys
import numpy as np
print(np.__version__)

vcffile = "subdata.vcf"
```

```python
import ipyrad.analysis as ipa
pca = ipa.pca(vcffile)
pca.plot()
```

```python
mask = pca.pcs.values[:, 0] > 1000
print(pca.samples_vcforder[mask])
pca.remove_samples(pca.samples_vcforder[mask])
```

```python
pca.plot()
```

```python
pca.plot_pairwise_dist()
```

```python
import allel
import scipy.spatial
dist = allel.stats.pairwise_distance(pca.genotypes.to_n_alt(), metric="euclidean")
scipy.spatial.distance.squareform(dist)
```

```python
import scipy.spatial
square = scipy.spatial.distance.squareform(dist)
#print(square)
#square.tofile("/tmp/lanei.txt", sep="\t")
with open("/tmp/lanei.txt" , 'wb') as f:
    np.savetxt(f, square, delimiter=' ', newline='\n', header='', footer='', fmt='%s')
```
