```python
io5 = h5py.File(data3, 'r')
dbnames = [i.decode() for i in io5["snps"].attrs["names"]]
print(io5["snps"].shape)
```

```python
idx = dbnames.index(imap3["G2"][0])
g2_snps = io5["snps"][idx, :]
missing_sites = set(np.where(g2_snps == 78)[0])

def shared_missing(group):
    missing = []
    for samp in imap3[group]:
        idx = dbnames.index(samp)
        #print("{} {}".format(idx, samp))
        samp_snps = io5["snps"][idx, :]
        msites = set(np.where(samp_snps == 78)[0])
        shared = missing_sites.intersection(msites)
        missing.append(len(shared))
    return missing

for group in imap3.keys():
    if group == "G2":
        continue
    missing = shared_missing(group)
    print(group)
    print(np.mean(missing))
    print(np.std(missing))
    
```
